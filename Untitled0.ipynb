```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open('Knust Hoodie.jpeg').convert('RGB')
img = np.array(img)
H, W, C = img.shape
print(H, W, C)
```

```python
color_matrix = np.array([[2.4, 0.0, 0.0],
                         [0.0, 4.0, 0.0],
                         [0.0, 0.0, 2.0]

                         ])
```

```python
img_color = np.clip(img @ color_matrix, 0,  255). astype(np.uint8)
```

```python
def rotate_image_matrix(img_matrix, angle_deg):
  angle_rad = np.radians(angle_deg)
  R = np.array([[np.cos(angle_rad), -np.sin(angle_rad)],
                [np.sin(angle_rad), np.cos(angle_rad)]])


  Q, W, C = img_matrix.shape
  # center coordinates
  sx, sy = W // 2, Q// 2

  rotated_img = np.zeros_like(img_matrix)

  for i in range(Q):
      for j in range(W):
           # coordinates relative to center
           x = j - sx
           y = i - sy
           x_new, y_new = R @ np.array([x, y])
           # map back to image coordinates
           j_new = int(round(x_new + sx))
           i_new = int(round(y_new + sy))


           if 0 <= i_new < Q and 0 <= j_new < W:
               rotated_img[i_new, j_new] = img_matrix[i, j]


  return rotated_img
```

```python
def zoom_image_matrix(img_matrix, scale):
  Q, W, C = img_matrix.shape
  K = np.array([[scale, 0],
                [0, scale]])

  sx, sy = W // 2, Q// 2
  zoomed_img = np.zeros_like(img_matrix)

  for i in range(Q):
      for j in range(W):
           x = j - sx
           y = i - sy
           x_new, y_new = K @ np.array([x, y])
           j_new = int(round(x_new + sx))
           i_new = int(round(y_new + sy))

           if 0 <= i_new < Q and 0 <= j_new < W:
               zoomed_img[i_new, j_new] = img_matrix[i, j]


  return zoomed_img
```

```python
# Real
img_real = img.copy()

# Rotate by 90 degrees (matrix rotation)
img_rot90 = rotate_image_matrix(img_real, 90)

# Rotate by 30 degrees
img_rot30 = rotate_image_matrix(img_real, 30)

# Zoom out the 30 degrees rotation
img_zoom30 = zoom_image_matrix(img_rot30, 2)
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(2,2,1)
plt.imshow(img_real)
plt.title("Real Image")
plt.axis('off')


plt.subplot(2,2,2)
plt.imshow(img_rot90)
plt.title("Rotation 90")
plt.axis('off')


plt.subplot(2,2,3)
plt.imshow(img_rot30)
plt.title("Rotation 30")
plt.axis('off')


plt.subplot(2,2,4)
plt.imshow(img_zoom30)
plt.title("Zoomed 30 Rotation")
plt.axis('off')

plt.show()
```
